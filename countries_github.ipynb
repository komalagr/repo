```python
import matplotlib.pyplot as plt
import pandas as pd
from pylab import *
```

```python
countries=pd.read_csv("countries.csv")
countries.head(10)
```

```python
countries.info()
```

# Compare population rate of Finland vs Sweden

```python
finland=countries[countries['country']=='Finland'] # get population data for Finland country
```

```python
sweden=countries[countries['country']=='Sweden'] # get population data for Sweden country
```

```python
x=sweden['year']
y=sweden['population']
z=finland['population']
plt.plot(x,y/10**6)
plt.plot(x,z/10**6)
plt.title("year - population chart")
plt.xlabel("year")
plt.ylabel('population')
plt.legend(['sweden','finland'])
```

# Function to Compare poultaion data of all countries for input year

```python
def year_data(year):
    population_data=countries[countries['year']==year] # get data only for those rows for the input year
    population_data[['year','population']]
    x_axis=population_data['country'] # list countries on x axis
    y_axis=population_data['population'] # lists 1952s population on y axis
    plt.figure(figsize=(50,25)) # chart would be bigger
    plt.bar(x_axis,y_axis,width=1) # bar chart 
    plt.xticks(rotation=70)
    plt.show()

year_number=int(input("Enter a year "))
year_data(year_number)
    
```
