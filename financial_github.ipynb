```python
def nasdaq(dataset):
    dataset.sort_values(by=['date'],ascending=True,inplace=True)
    

nasdaq(google1)
    
```

```python
def nasdaq(*args): # if we dont number of arguments
    for dataset in args:
        dataset.sort_values(by=['date'],ascending=True,inplace=True)
        
nasdaq(microsoft1,google1,amazon1,apple1)
```
